```python
#proses cleansing
def cleansing(text):
    string = text.lower() #menjadi huruf keci
    string = re.sub(r'[^a-zA-Z0-9]', ' ', string) #remove selain huruf
    string = re.sub(r'\d', '', string) # remove number

    return string
```

```python
#melihat hasil cleansing
kamusalay = pd.read_csv('new_kamusalay.csv', encoding = 'latin-1')
kamusalay = kamusalay['anak jakarta asyik asyik'].apply(cleansing)
kamusalay
```

```python
#menambahkan kolom hasil cleansing
kamusalay = pd.read_csv('new_kamusalay.csv', encoding='latin-1')
kamusalay['anak jakarta asyik asyik baru'] = kamusalay['anak jakarta asyik asyik'].apply(cleansing)
kamusalay
```

```python
#menghapus kolom lama sebelum cleansing
kamusalay = kamusalay.drop('anak jakarta asyik asyik', axis=1)
kamusalay
```

```python
#menyimpan data new_kamus alay baru yg sudah di cleansing
kamusalay.to_csv('new_kamusalay_cleansing.csv')
```

cleansig

```python
#proses cleansing
def cleansing(text):
    string = text.lower() #mengubah huruf kapital menjadi huruf kecil
    string = re.sub(r'\burl\b', '', string) #menghilangkan kata URL 
    string = re.sub(r'\\[^\s]+', '', string) #menghilankan kata yang dimulai dengan \
    string = re.sub(r'rt|wk|[^\w\s]', '', string) #menghilangkan kata wkwkwkw
    string = re.sub(r'[^a-zA-Z0-9]', ' ', string) #remove selain huruf
    string = re.sub(r'rt|user|[^\w\s]', '', string) #remove kata usename dan dan kata selanjutnya
    string = re.sub(r'\d', '', string) # remove number
    string = re.sub('((www\.[^\s]+)|(https?://[^\s]+)|(http?://[^\s]+))',' ', string) #remove link url
    string = re.sub('  +', ' ', string) # Remove extra space
    
    return string



   
```

```python
#melihat hasil cleansing
data = pd.read_csv('data.csv', encoding='latin-1')
data = data['Tweet'].apply(cleansing)
data
```

```python
#menambahkan kolom hasil cleansing
data = pd.read_csv('data.csv', encoding='latin-1')
data['Tweet After Cleansing'] = data['Tweet'].apply(cleansing)
data
```

```python
#menghapus kolom sebelum cleansing
data = data.drop('Tweet', axis=1)
data
```
